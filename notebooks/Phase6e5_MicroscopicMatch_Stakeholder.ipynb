## What's Λ^emerg?

When you integrate out the microscopic fermion theory at a high
cutoff `Λ_UV`, the leading term you generate is dimension-4 ---
i.e., it scales as `Λ_UV⁴`. That's the cosmological constant.
There's a known coefficient (Christensen-Duff Dirac, 1979):

$$\Lambda^{\mathrm{emerg}} = \frac{N_f}{4\pi^2}\,\Lambda_{\mathrm{UV}}^4.$$

Plug in the natural numbers --- N_f = 16 (the Standard-Model
Dirac fermion count, more or less) and Λ_UV = M_Planck = 10¹⁹
GeV --- and you get a number that's *much* larger than what we
see in the sky.

```python
from src.core.constants import MICRO_MACRO_PARAMS
from src.core.formulas import lambda_emerg_microscopic
import math
M_Pl = MICRO_MACRO_PARAMS['M_PLANCK_GEV']
Lambda_obs = MICRO_MACRO_PARAMS['LAMBDA_OBSERVED_GEV4']
le = lambda_emerg_microscopic(M_Pl, 16)
ratio = le / Lambda_obs
print(f'Microscopic prediction: Λ^emerg ≃ {le:.2e} GeV⁴')
print(f'Planck-2018 observed:    Λ_obs   ≃ {Lambda_obs:.2e} GeV⁴')
print(f'Ratio: ~10^{math.log10(ratio):.0f} — about 122 orders too big.')
```

## Could we fine-tune our way out?

What if we pick smaller `Λ_UV` or smaller `N_f`? In principle
yes --- mathematically you can choose `Λ_UV ≃ 4.5 meV` (4.5
millielectronvolts), and at that cutoff the heat-kernel
prediction *does* match Λ_obs. But that 'cutoff' is only ~50
times larger than the energy of a microwave photon. No natural
UV-completion theory has its cutoff there. So the resolution is
*formally* possible but *physically* unnatural: it requires
throwing away the Standard Model.

Wave 5 makes this exact statement: the verdict band
`cc_resolved` is real, but it sits at a far-sub-electroweak
cutoff that no high-energy theory inhabits.

```python
from src.micro_macro_match import (
    natural_parameter_scan, resolution_locus_diagnostic,
)
from collections import Counter
scan = natural_parameter_scan(
    N_f_values=[1, 4, 16, 100], n_lambda_uv_points=8
)
natural = [a for a in scan if a.natural]
verdict_counts = Counter(a.verdict for a in natural)
print('Natural-band verdict tally:')
for v, c in verdict_counts.items():
    print(f'  {v}: {c}')
print()
locus = resolution_locus_diagnostic(16.0)
print(f'Resolution-locus cutoff: Λ_UV ≃ {locus.Lambda_UV_GeV:.2e} GeV (a few meV — far sub-electroweak).')
print(f'Verdict at locus: {locus.verdict}; natural={locus.natural}')
```
