## A naming caveat up front

This note compares two distinct entropy notions that share the $A/(4 G_N)$ leading term:

- $S_{BH}$ — black-hole *thermodynamic* entropy (Bekenstein-Hawking, Kaul-Majumdar microscopic counting).
- $S_{\rm ent}$ — *entanglement* entropy of a boundary CFT region (Ryu-Takayanagi, Calabrese-Cardy, Casini-Huerta).

In AdS/CFT they are deeply connected (Ryu-Takayanagi *equates* boundary entanglement entropy with bulk minimal-surface area divided by 4G_N), but they are not the same object. Phase 6a Wave 3's Kaul-Majumdar derivation gave a microscopic count of $S_{BH}$ for the project's horizon, with a $-3/2$ log correction. The classical RT *formula* gives just $A/(4 G_N)$. Whether RT is taken as a candidate $S_{BH}$ or as $S_{\rm ent}$, comparing it to Kaul-Majumdar's microscopic count exposes the same structural fact: leading-order $A/(4 G_N)$ alone cannot match the microscopic theory, and the gap in our setup is exactly $(3/2)\log(A/(4 G_N))$.

## The Bekenstein-Hawking starting point

In 1973, Bekenstein argued black holes have entropy. In 1974, Hawking computed it: $S_{BH} = A / (4 G_N)$ (in natural units). One quarter the area, in Planck units. This is the most-cited equation in quantum gravity. It's also the *leading* term: the asymptotic-area-law part. It says nothing about what microscopic theory underlies it.

## The log correction

Different microscopic theories of quantum gravity predict different *subleading* corrections — usually logarithms of the area:

$$S_{BH} = \frac{A}{4 G_N} + \alpha \log\!\left(\frac{A}{4 G_N}\right) + \dots$$

The coefficient $\alpha$ depends on the microscopic theory. It is *not* universal across approaches:

- **Kaul–Majumdar (2000)** counted SU(2)$_k$ Chern-Simons edge states on the horizon: $\alpha = -3/2$.
- **Sen (2012)** computed the same quantity using string-theory loop amplitudes for 4D Schwarzschild: $\alpha = 77/45 \approx +1.71$.

Same physical entropy, two different microscopic predictions, gap of $\approx 3.21$. *Not* a numerical accident — a genuine divergence between theories.

## Ryu-Takayanagi and the missing log

Ryu and Takayanagi (2006) made *boundary entanglement entropy* geometric in AdS/CFT: $S_{\rm ent} = (\text{minimal-surface area})/(4 G_N)$. No log term at leading order. *That cannot be a complete microscopic match* if Kaul-Majumdar's $-3/2$ log correction governs the corresponding $S_{BH}$.

## Universal 2D-CFT entanglement-entropy bound

For 2D conformal field theories, Holzhey–Larsen–Wilczek (1994) and Calabrese–Cardy (2004) provided the *universal* upper bound on entanglement entropy of a finite interval: $S_{\rm ent}(L) \leq (c/3)\log(L/\epsilon)$, where $c$ is the central charge and $\epsilon$ a UV cutoff. Casini–Huerta (2009) gave the rigorous free-QFT specialization. Earlier project documents — and the Lean predicate `H_CasiniHuerta_Bound_Valid` — used "Casini–Huerta" as shorthand; we keep that name for the predicate but cite the universal source correctly here.

## What this note shows

Phase 6a Wave 3 derived the Kaul-Majumdar $-3/2$ coefficient from the project's microscopic horizon boundary condition. This note formalizes the *structural consequence*: classical Ryu-Takayanagi (no log term) and Kaul-Majumdar ($-3/2 \log$) cannot simultaneously describe the same entropy quantity. They agree on a single knife-edge: $A = 4 G_N$ exactly, where $\log(1) = 0$ kills the discrepancy. Anywhere else, they differ by $(3/2)\log(A/(4 G_N))$.

## 3. Why log coefficients aren't universal

Both Kaul-Majumdar and Sen claim to compute *the* universal log coefficient for black-hole entropy. They get different answers. This is not because one of them is wrong — they're computing different things implicitly:

- Kaul-Majumdar: SU(2)$_k$ Chern-Simons edge-mode counting on a 2D horizon.
- Sen: string-theory amplitude calculation for 4D Schwarzschild specifically.

Different *regularizations*, different *boundary conditions*, different *microscopic dictionaries*. The log coefficient $\alpha$ depends on these choices. A theory that gets the leading area term right (BH entropy) can still get the subleading log wrong — and *which* log is right is the very question microscopic theories are trying to answer.

```python
print(f'  Kaul-Majumdar log coefficient   = -3/2 = {-1.5:.4f}')
print(f'  Sen 4D Schwarzschild coefficient = {sen_4d_log_coeff():.4f}    (= 77/45 ≈ +1.711)')
print(f'  gap                              = {SEN_KAUL_MAJUMDAR_COEFF_GAP:.4f}    (= 289/90 ≈ 3.211)')
print()
print('  → Sen and Kaul-Majumdar are not refuting each other; they are answering')
print('    different microscopic questions. The point is: the log coefficient is a')
print('    SIGNATURE of microscopic physics, not a universal constant.')
```

## 4. Universal 2D-CFT entanglement-entropy bound

For 2D conformal field theory, Holzhey–Larsen–Wilczek (1994) and Calabrese–Cardy (2004) proved the universal upper bound on entanglement entropy of an interval of length $L$:
$$S_{\rm ent}(L) \leq \frac{c}{3}\log\!\left(\frac{L}{\epsilon}\right),$$
where $c$ is the CFT's central charge and $\epsilon$ a UV cutoff. The bound is *tight* — saturated by free CFTs (rigorous case treated by Casini–Huerta 2009).

Compute the saturated entropy at $L = 1$ across canonical 2D CFTs:

```python
UV = 1e-3
L = 1.0
for label, c in [
    ('Ising',                 0.5),
    ('Tricritical Ising',     0.7),
    ('3-state Potts',         0.8),
    ('Free boson',            1.0),
]:
    S = saturated_ch_entropy(L, c, UV)
    print(f'  {label:24s}  c = {c:.2f}    S_ent(L=1) saturated = {S:.4f}')

print()
print('  → Higher central charge = larger saturated entropy at the same L.')
print('    The bound is *log* — different microscopic regularizations would all')
print('    have to respect this universal envelope.')
```
