## What does "negative sound speed squared" mean?

A perfect fluid's sound speed squared $c_s^2$ tells you how fast
small density perturbations propagate. For a normal fluid (water,
photons, dust), $c_s^2 > 0$ and perturbations oscillate as sound
waves. For a relativistic fluid (radiation), $c_s^2 = 1/3$. For a
cosmological constant, $c_s^2 = 1$.

The Phase 5y vestigial EOS (derived from the H4 Volovik-style
analogy) gives a closed-form expression for $c_s^2$ as a function
of the dimensionless temperature $\tau$:

$$c_s^2(\tau) = -\frac{1 - \tau^2}{3 - 5\tau^2}.$$

At $\tau = 0$ (the deep-vestigial limit, which is the regime
relevant for late-time DESI cosmology), this evaluates to
$c_s^2(0) = -1/3$. **Negative.**

## Why does the sign of $c_s^2$ matter for the CMB?

The linear scalar perturbation $\delta$ at sub-horizon scales obeys
a wave equation analogous to the simple harmonic oscillator:
$\ddot{\delta} + (\textrm{Hubble damping}) + c_s^2 k^2 \delta = 0.$

The sign of $c_s^2$ determines whether you get **oscillations**
($c_s^2 > 0$, like a normal pendulum) or **exponential growth**
($c_s^2 < 0$, like a pendulum hanging *up* from a pivot — unstable
equilibrium).

For the CMB, the angular power spectrum amplitude $|C_\ell|^2$
is the squared evolution of the perturbation since recombination.
Cosines stay bounded; cosh's diverge at all scales.
