## The Problem: Why Can't We Put Chirality on a Lattice?

The Nielsen-Ninomiya theorem (1981) says: if you try to put a single
left-handed fermion on a lattice with the obvious symmetry properties,
you always get an unwanted right-handed partner (a "doubler"). This is
the chirality wall.

In 2026, Gioia and Thorngren found a way around it: use a different kind
of symmetry that isn't "on-site" (it connects neighboring lattice points).

## The Solution: One Weyl Fermion, Machine-Verified

The GT construction produces exactly **one** massless chiral fermion.
The "Wilson mass" function gaps all the unwanted doublers while preserving
the one we want.

```python
# Verify: exactly 1 Weyl node on lattices of different sizes
for L in [4, 8, 16]:
    k = brillouin_zone(L)
    nodes = find_weyl_nodes(k)
    print(f'Lattice size L={L}: {len(nodes)} Weyl node (chiral!)')
```

## The Proof: [H, Q_A] = 0
