## What Are Quantum Groups?

A **quantum group** is a continuous deformation of an ordinary symmetry group. The deformation
is controlled by a parameter $q$:

- At $q = 1$: you get back the ordinary symmetry (classical limit)
- At $q \neq 1$: the symmetry is "stretched" -- multiplication becomes non-commutative in a controlled way
- At $q = $ root of unity: something remarkable happens -- the representation theory becomes finite and produces **topological quantum field theories**

The simplest quantum group, $U_q(\mathfrak{sl}_2)$, deforms the rotation group. It has four
generators (raise, lower, and two Cartan elements) satisfying relations that reduce to
ordinary angular momentum algebra at $q = 1$.

**Analogy:** Think of $q$ as a "tuning knob." At one setting ($q=1$), you have ordinary
rotations. As you turn the knob, the rotations become "quantum" -- they remember their
order (AB is not the same as BA). At special settings (roots of unity), the system
crystallizes into a finite set of topological states.

## q-Deforming the Onsager Algebra

In earlier waves, we formalized the **Onsager algebra** -- the symmetry behind exactly solvable
models like the 2D Ising model. It is defined by the Dolan-Grady relations, which involve a
"triple commutator" with coefficient 3 and a characteristic constant 16.

**q-deformation** replaces the integer 3 with a $q$-integer $[3]_q$:

$$[3]_q = q^2 + 1 + q^{-2}$$

At $q = 1$, this gives $1 + 1 + 1 = 3$, recovering the original Onsager algebra. This is
the entry point into the world of quantum groups.

```python
# The q-integer: a "stretched" version of ordinary counting
print('q-integers: counting with a twist')
print('=' * 45)
print(f'{"n":>3}  {"[n]_1 (classical)":>18}  {"[n]_2 (stretched)":>18}')
print('-' * 45)
for n in range(6):
    classical = q_integer(n, 1)
    stretched = q_integer(n, 2)
    print(f'{n:>3}  {classical:>18.0f}  {stretched:>18.1f}')

print(f'\nAt q=1, [n]_q = n (ordinary integers).')
print(f'At q=2, the integers are "stretched" -- they grow faster.')
```

## Why This Is a First

No proof assistant -- not Lean, not Coq, not Isabelle, not Agda -- has ever defined a
**quantum group**. Our `Uqsl2.lean` is the first.

What makes this possible:
- Lean 4's Mathlib library already has `FreeAlgebra` (free algebras) and `RingQuot` (quotient rings)
- We follow the exact same pattern used for `TensorAlgebra` and `ExteriorAlgebra` in Mathlib
- The result: $U_q(\mathfrak{sl}_2)$ is **defined** (constructed from existing math), not **axiomatized** (assumed to exist)

This matters because defined objects are automatically consistent -- they inherit all the properties
of the underlying mathematical structures. Axiomatized objects can harbor contradictions.
