## From Rank-1 to Rank-2: Why It Matters

Going from SU(2) to SU(3) is not just "one more" — it is a qualitative leap.
SU(2) has a single type of charge (like electric charge: positive or negative).
SU(3) has two independent types of charge (like the "color" charges of quarks:
red, green, blue).

The mathematical complexity grows accordingly: SU(2) needs 4 generators and
simple relations. SU(3) needs 8 generators and 21 relations, including the
cubic Serre relations that have no analogue in rank-1.

```python
print("Comparison: SU(2) vs SU(3) quantum groups")
print(f"{'':20s} {'SU(2)':>10s} {'SU(3)':>10s}")
print(f"{'Rank':20s} {'1':>10s} {'2':>10s}")
print(f"{'Generators':20s} {'4':>10s} {'8':>10s}")
print(f"{'Relations':20s} {'7':>10s} {'21':>10s}")
print(f"{'Serre relations':20s} {'0':>10s} {'4':>10s}")
print(f"{'Cartan matrix':20s} {'1x1':>10s} {'2x2':>10s}")
print()
print("At level k, the number of anyon types also grows faster:")
print(f"{'Level k':>8s} {'SU(2)_k':>10s} {'SU(3)_k':>10s}")
for k in [1, 2, 3, 4]:
    print(f"{k:>8d} {k+1:>10d} {(k+1)*(k+2)//2:>10d}")
```

## SU(3) Anyons Contain a Quantum Computer

SU(3) at level 2 has 6 types of anyons. Among them is an object called "tau"
that satisfies the **Fibonacci fusion rule**: tau x tau = 1 + tau.

This is significant because the Fibonacci anyon is **universal for quantum
computation** — any quantum algorithm can be executed by braiding Fibonacci
anyons. The fact that it appears naturally inside SU(3)_2 means that any
physical system realizing SU(3) level-2 Chern-Simons theory automatically
contains a universal quantum computer.

```python
phi = (1 + np.sqrt(5)) / 2

print("SU(3)_2 anyon content:")
print("  6 simple objects: vacuum(1), fundamental(f), conjugate(f-bar),")
print("                    symmetric(s), conjugate-symmetric(s-bar), adjoint(tau)")
print()
print("Fibonacci subcategory {1, tau}:")
print(f"  tau x tau = 1 + tau")
print(f"  Quantum dimension of tau = golden ratio = {phi:.6f}")
print(f"  This subcategory is UNIVERSAL for quantum computation.")
print()
print("Proved in Lean 4: su3k2_fibonacci_subcategory (SU3kFusion.lean)")
```
