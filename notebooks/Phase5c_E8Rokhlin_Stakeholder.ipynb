## The $E_8$ Lattice: Nature's Most Perfect Pattern

Imagine stacking oranges as efficiently as possible, but in 8 dimensions instead of 3.
The most efficient packing is the **$E_8$ lattice** -- and it has remarkable properties:

- **Even:** Every vector has even length-squared (like a checkerboard pattern)
- **Unimodular:** The lattice is its own dual (no "holes" between lattice points)
- **Positive definite:** All directions have positive length (no null vectors)

These properties are captured by an 8x8 matrix called the **Cartan matrix**. The matrix
has 2 on the diagonal (even) and determinant 1 (unimodular).

The **signature** of this matrix -- the difference between positive and negative eigenvalues --
is $\sigma = 8$. This is the crucial number.

```python
# Build and verify the E8 matrix
E8 = np.array([
    [ 2,  0, -1,  0,  0,  0,  0,  0],
    [ 0,  2,  0, -1,  0,  0,  0,  0],
    [-1,  0,  2, -1,  0,  0,  0,  0],
    [ 0, -1, -1,  2, -1,  0,  0,  0],
    [ 0,  0,  0, -1,  2, -1,  0,  0],
    [ 0,  0,  0,  0, -1,  2, -1,  0],
    [ 0,  0,  0,  0,  0, -1,  2, -1],
    [ 0,  0,  0,  0,  0,  0, -1,  2],
])

eigenvalues = np.sort(np.linalg.eigvalsh(E8))
print('The E8 Cartan matrix -- key properties:')
print(f'  Determinant = {round(np.linalg.det(E8))} (unimodular: the lattice is its own dual)')
print(f'  All eigenvalues positive: {np.all(eigenvalues > 0)} (positive definite)')
print(f'  Signature = {int(np.sum(eigenvalues > 0))} (all 8 eigenvalues are positive)')
```

## The Algebra-Topology Gap: 8 vs 16

Here is the central puzzle:

**Algebra says:** The signature of any even, unimodular lattice must be divisible by **8**.
This is a theorem of Serre (1970), proved using number theory. The $E_8$ lattice achieves
the minimum: $\sigma = 8$.

**Topology says:** If this lattice arises as the intersection form of a smooth 4-dimensional
space (manifold), then the signature must be divisible by **16**. This is Rokhlin's theorem
(1952).

So $E_8$ (with $\sigma = 8$) *cannot* come from a smooth 4-manifold. Freedman proved in
1982 that a *topological* (but not smooth) $E_8$ manifold does exist.

**The gap:** Between 8 and 16 is a factor of 2 that has no algebraic explanation.
It comes from the **smoothness of space** -- the requirement that you can do calculus,
not just topology.

```python
# Visualize the gap
sigma_E8 = 8

print('The 8 vs 16 gap:')
print('=' * 50)
print(f'  E8 signature:     sigma = {sigma_E8}')
print(f'  sigma mod 8  = {sigma_E8 % 8}   -> Algebra: ALLOWED')
print(f'  sigma mod 16 = {sigma_E8 % 16}   -> Smooth topology: FORBIDDEN')
print()
print(f'  To satisfy Rokhlin, need sigma = 16:')
print(f'  E8 + E8 has sigma = 16')
print(f'  16 mod 16 = {16 % 16}   -> Smooth topology: ALLOWED')
print()
print(f'  Gap factor: 16 / 8 = 2')
print(f'  This factor of 2 is invisible to algebra.')
print(f'  It comes from the topology of smooth 4-manifolds.')
```
