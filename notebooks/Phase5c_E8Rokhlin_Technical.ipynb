## 1. The $E_8$ Cartan Matrix

The $E_8$ root system is the unique even, unimodular, positive-definite lattice in
8 dimensions. Its Cartan matrix $A$ encodes the inner products of simple roots:

$$A_{ij} = \frac{2 \langle \alpha_i, \alpha_j \rangle}{\langle \alpha_j, \alpha_j \rangle}$$

Properties to verify:
- $\det(A) = 1$ (unimodularity)
- $A_{ii} = 2$ for all $i$ (even lattice)
- $A$ is positive definite (all eigenvalues $> 0$)
- Signature $\sigma(E_8) = 8$ (all eigenvalues positive, rank 8)

```python
# E8 Cartan matrix
E8 = np.array([
    [ 2,  0, -1,  0,  0,  0,  0,  0],
    [ 0,  2,  0, -1,  0,  0,  0,  0],
    [-1,  0,  2, -1,  0,  0,  0,  0],
    [ 0, -1, -1,  2, -1,  0,  0,  0],
    [ 0,  0,  0, -1,  2, -1,  0,  0],
    [ 0,  0,  0,  0, -1,  2, -1,  0],
    [ 0,  0,  0,  0,  0, -1,  2, -1],
    [ 0,  0,  0,  0,  0,  0, -1,  2],
])

det_E8 = round(np.linalg.det(E8))
eigenvalues = np.sort(np.linalg.eigvalsh(E8))
all_positive = np.all(eigenvalues > 0)
diag_check = np.all(np.diag(E8) == 2)
even_check = np.all(E8 % 2 == 0) or np.all(np.diag(E8) % 2 == 0)
symmetric = np.allclose(E8, E8.T)
n_pos = np.sum(eigenvalues > 0)
n_neg = np.sum(eigenvalues < 0)
sigma = int(n_pos - n_neg)

print('E8 Cartan matrix verification:')
print('=' * 50)
print(f'  Determinant:       {det_E8} (expected 1 for unimodular)')
print(f'  Symmetric:         {symmetric}')
print(f'  Diagonal = 2:      {diag_check} (even lattice)')
print(f'  Positive definite: {all_positive}')
print(f'  Eigenvalues:       {np.round(eigenvalues, 4)}')
print(f'  Signature:         sigma = {n_pos} - {n_neg} = {sigma}')

assert det_E8 == 1, 'E8 not unimodular!'
assert all_positive, 'E8 not positive definite!'
assert sigma == 8, 'E8 signature is not 8!'
```

## 2. Algebraic Serre Theorem: $8 | \sigma$

**Theorem (Serre, 1970):** Every even, unimodular lattice in $\mathbb{R}^n$ has
$n \equiv 0 \pmod{8}$, and the signature satisfies $\sigma \equiv 0 \pmod{8}$.

This is a purely algebraic result -- it follows from properties of even unimodular
forms over $\mathbb{Z}$. The $E_8$ lattice saturates this bound with $\sigma = 8$.

**Proof sketch:** An even unimodular form $Q$ over $\mathbb{Z}$ has $Q \otimes \mathbb{F}_2$
as a Type II code over $\mathbb{F}_2$. The length of such codes satisfies $n \equiv 0 \pmod{8}$.
Combined with $|\sigma| \leq n$, this gives $8 | \sigma$.

```python
# Verify Serre's theorem: sigma = 0 mod 8 for even unimodular lattices
# Build various even unimodular forms and check

# Hyperbolic plane H = ((0,1),(1,0)), signature 0
H = np.array([[0, 1], [1, 0]])
sig_H = int(np.sum(np.linalg.eigvalsh(H) > 0) - np.sum(np.linalg.eigvalsh(H) < 0))

# E8 (+) E8: signature 16
E8E8 = np.block([[E8, np.zeros((8,8))], [np.zeros((8,8)), E8]])
sig_E8E8 = 16  # Two copies of E8

# E8 (+) H: signature 8 (not unimodular as direct sum, but E8 perp H has det = -1)
# The standard indefinite unimodular form: E8 (+) H
# Actually the classification says: even unimodular <=> aE8 (+) bH for some a,b

print('Even unimodular lattice classification (Milnor-Husemoller):')
print('=' * 60)
print(f'  E8:     sigma = {sigma:>3},  sigma mod 8 = {sigma % 8}')
print(f'  H:      sigma = {sig_H:>3},  sigma mod 8 = {sig_H % 8}')
print(f'  E8+E8:  sigma = {sig_E8E8:>3}, sigma mod 8 = {sig_E8E8 % 8}')

# General classification: sigma = 8(a - b) where a = # of E8, b = # of (-E8)
print(f'\nGeneral form: any even unimodular lattice has sigma = 8a - 8b = 8(a-b)')
print(f'where a = number of E8 summands, b = number of -E8 summands.')
print(f'The hyperbolic planes H contribute 0 to sigma.')
print(f'\nThis is ALGEBRAIC: it holds for any even unimodular form over Z.')
```

## 3. Why 16 Requires Topology: Rokhlin's Theorem

**Rokhlin's theorem (1952):** If $M$ is a closed, oriented, smooth, spin 4-manifold,
then $\sigma(M) \equiv 0 \pmod{16}$.

The key word is **smooth**. Algebra gives $8 | \sigma$. Topology adds the extra factor of 2.

**Freedman's $E_8$ manifold (1982):** There exists a closed, oriented, *topological*
4-manifold with intersection form $E_8$ (hence $\sigma = 8$). But Rokhlin says $16 | \sigma$
for smooth manifolds, so:

> Freedman's $E_8$ manifold is not smoothable.

This is the **algebra-topology gap**: algebra permits $\sigma = 8$, but smooth topology
requires $\sigma = 16$. The extra factor of 2 comes from the interplay of spin structure
and smooth structure -- it has no purely algebraic explanation.

```python
# The algebra-topology gap
print('The Algebra-Topology Gap:')
print('=' * 60)
print()
print('  Algebraic constraint (Serre):    sigma = 0 mod 8')
print('  Topological constraint (Rokhlin): sigma = 0 mod 16')
print()
print('  E8 lattice:   sigma = 8')
print('    -> 8 mod 8  = 0  (satisfies algebra)')
print('    -> 8 mod 16 = 8  (VIOLATES smooth topology)')
print()
print('  E8 + E8:      sigma = 16')
print('    -> 16 mod 8  = 0  (satisfies algebra)')
print('    -> 16 mod 16 = 0  (satisfies smooth topology)')
print()
print('  Consequence: Freedman\'s E8 manifold (sigma=8) exists')
print('  topologically but CANNOT be made smooth.')
print()
print('  The gap factor: 16/8 = 2')
print('  This factor of 2 is PURELY topological -- no algebra can see it.')
```

## 4. The Hyperbolic Plane and Indefinite Classification

The **hyperbolic plane** $H = \begin{pmatrix} 0 & 1 \\ 1 & 0 \end{pmatrix}$ has
eigenvalues $\pm 1$, so $\sigma(H) = 0$. It is even (diagonal entries are 0) and
unimodular ($\det = -1$, but as an intersection form on a 4-manifold, we need $|\det| = 1$).

**Classification theorem (Milnor-Husemoller):** Every indefinite, even, unimodular
lattice is isomorphic to $a \cdot E_8 \oplus b \cdot (-E_8) \oplus c \cdot H$ for
some non-negative integers $a, b, c$. The signature is $\sigma = 8(a - b)$.

This means the signature of any such lattice is automatically divisible by 8, confirming
Serre's theorem. The classification fails for positive-definite forms (there are finitely
many in each dimension, growing rapidly).

```python
# Hyperbolic plane properties
H = np.array([[0, 1], [1, 0]])
eigs_H = np.linalg.eigvalsh(H)

print('Hyperbolic plane H:')
print(f'  Matrix: [[0, 1], [1, 0]]')
print(f'  Eigenvalues: {np.round(eigs_H, 4)}')
print(f'  Determinant: {round(np.linalg.det(H))}')
print(f'  Signature: {int(np.sum(eigs_H > 0) - np.sum(eigs_H < 0))}')
print(f'  Even: diagonal entries are {np.diag(H)} (all even)')

# Build example: E8 + (-E8) + H (sigma = 0, rank = 18)
# Show the classification pattern
print(f'\nClassification sigma = 8(a - b):')
print(f'{"a (E8)":>8}  {"b (-E8)":>8}  {"c (H)":>8}  {"rank":>6}  {"sigma":>6}')
print('-' * 45)
examples = [(1, 0, 0), (0, 1, 0), (1, 1, 1), (2, 0, 0), (2, 1, 0), (1, 0, 3)]
for a, b, c in examples:
    rank = 8*a + 8*b + 2*c
    sig = 8*(a - b)
    print(f'{a:>8}  {b:>8}  {c:>8}  {rank:>6}  {sig:>6}')
```
