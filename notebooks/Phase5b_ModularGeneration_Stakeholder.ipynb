## The Question

The Standard Model of particle physics has exactly three "generations" (copies) of its fundamental particles. Electrons, muons, and taus are three versions of the same particle with increasing mass. Why three? Not two, not four?

## The Answer: 24 ÷ 8 = 3

We've formally verified that the number of generations must be a **multiple of 3**, and traced this to a beautiful equation:

- **24** comes from pure mathematics — the Dedekind eta function, a 19th-century object from number theory
- **8** comes from pure physics — counting the Standard Model's fundamental particles (16 particles, each contributing 1/2)
- **3** = 24/8 is the generation constraint

## Step 1: Counting Particles

Each generation of the Standard Model has 16 fundamental particle components (quarks in 3 colors × 2 weak charges, leptons, and a right-handed neutrino).

```python
print('Standard Model particles per generation:')
print('=' * 45)
total = 0
for name, f in SM_FERMION_DATA.items():
    print(f'  {name:15s}: {f["components"]} components')
    total += f['components']
print(f'{"":15s}  --------')
print(f'{"Total":15s}: {total} components')
print(f'\nCentral charge: {total} × (1/2) = {total/2} per generation')
```

## Step 2: The Mathematical Constraint

A deep result from mathematical physics: for the theory to be self-consistent, the central charge must be divisible by 24. With a central charge of 8 per generation:

$$8 \times N_{\text{generations}} \equiv 0 \pmod{24}$$

This means $N_{\text{generations}}$ must be a multiple of 3.

```python
print('Testing each generation count:')
print(f'{"N_f":>5} {"c₋ = 8×N_f":>12} {"24 divides c₋?":>16} {"Allowed?":>10}')
print('-' * 48)
for n in range(1, 10):
    c = 8 * n
    divides = c % 24 == 0
    print(f'{n:>5} {c:>12} {str(divides):>16} {"✓" if divides else "✗":>10}')
```

This is a formal proof that right-handed neutrinos must exist — independent of the usual neutrino mass argument.
