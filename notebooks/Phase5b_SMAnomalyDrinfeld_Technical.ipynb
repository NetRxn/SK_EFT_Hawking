## 1. SM Fermion Z4 Charges

The discrete Z4 symmetry is generated by X = 5(B-L) - 4Y.

```
# viz-ref: fig_sm_fermion_z16_anomaly
for name, f in SM_FERMION_DATA.items():
    X = sm_z4_charge(f["B_minus_L"], f["Y"])
    comp = f["components"]
    print(f"{name:12s}: X={X} (odd={X%2==1}), components={comp}")
total_with = sum(f["components"] for f in SM_FERMION_DATA.values())
total_without = sum(f["components"] for n, f in SM_FERMION_DATA.items() if n != "nu_R_bar")
print(f"\nTotal with nu_R: {total_with}")
print(f"Total without nu_R: {total_without}")

fig_sm_fermion_z16_anomaly().show()
```

## 3. Generation Constraint

c_minus = 8 N_f, modular invariance requires c_minus = 0 mod 24, hence N_f = 0 mod 3.

```
# viz-ref: fig_sm_generation_constraint
for nf in range(1, 7):
    r = sm_generation_constraint(nf)
    satisfies = r["satisfies_generation_constraint"]
    mark = "***" if satisfies else "   "
    c_minus = r["c_minus"]
    mod24 = r["c_minus_mod24"]
    print(f"{mark} N_f={nf}: c_minus={c_minus}, mod24={mod24}, satisfies={satisfies}")

fig_sm_generation_constraint().show()
```
