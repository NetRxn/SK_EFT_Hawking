## Wave-by-Wave Results

### Wave 1: How Does the Hawking Effect Scale?

The Hawking effect depends on a parameter called "surface gravity" ($\kappa$) — how sharply the speed of sound changes at the acoustic horizon. Our EFT predicts two types of corrections to the Hawking temperature, and they scale differently with $\kappa$:

- **Dispersive correction** (from the lattice-like structure of the BEC): grows as $\kappa^2$
- **Dissipative correction** (from friction in the fluid): grows as $\kappa^1$

This different scaling is an experimental handle: by measuring the Hawking spectrum at multiple values of $\kappa$, experimenters can separate the two effects. The Heidelberg K-39 experiment, with its Feshbach-tunable scattering length, is best positioned for this test.

### Wave 2A: Polariton Systems — A Factor of 10 Billion

Polaritons are hybrid light-matter quasiparticles in semiconductor cavities. They form a Bose-Einstein condensate with a speed of sound a million times faster than atomic BECs, producing Hawking temperatures roughly $10^{10}\times$ higher.

The challenge: polaritons decay rapidly (the cavity leaks photons). For long-lifetime cavities ($\tau > 100$ ps), the cavity decay rate is small enough that our perturbative EFT treatment applies. For standard cavities ($\tau \sim 3$ ps), it does not.

```python
rows = []
for name, plat in POLARITON_PLATFORMS.items():
    rows.append({
        "Platform": name.replace("_", " "),
        "Cavity lifetime": f"{plat['tau_cav'] * 1e12:.0f} ps",
        "Hawking temp": f"{plat['T_H_K']:.2e} K",
        "EFT applicable?": "Yes" if plat['tier1_valid'] else "No",
        "Regime": plat['tier1_regime'],
    })

display(pd.DataFrame(rows))
```

### Wave 2B: Did Spacetime Split in Two?

Our Monte Carlo simulation of the Diakonov model for emergent gravity found evidence for a **split phase transition**: as coupling increases, the metric (which measures distances) orders before the tetrad (which defines local orientation). This means there could be a phase of spacetime where geometry exists but local frames do not — the "vestigial" phase predicted by Volovik.

This is a pilot study on small lattices. Confirming the split transition at large lattice sizes is a key target for future work.

### Waves 4A–4C: Mathematical Infrastructure for Topological Phases

String-net condensation (Levin-Wen) provides a lattice construction of topological quantum field theories. The mathematical input is a **fusion category** — a structure that describes how particles fuse together.

We built the first-ever formalization of fusion categories in a theorem prover (Lean 4), including:
- Pivotal and spherical categories (needed for consistent particle traces)
- Fusion rules and F-symbols (the "multiplication table" for particles)
- The Drinfeld double (how gauge symmetry emerges from fusion)

A key finding: the Drinfeld center $Z(\text{Vec}_G)$ is always non-chiral. This is the mathematical reason why gauge information cannot pass through the hydrodynamic boundary — the categorical foundation for the gauge erasure theorem from Phase 3.

```python
# Fusion category examples in plain language
rows = []
for name, ex in FUSION_EXAMPLES.items():
    D_sq = global_dimension_squared(ex['quantum_dims'])
    rows.append({
        "Category": name,
        "What it is": (
            f"Vector spaces graded by {ex['group']}" if ex.get('group')
            else "Non-group category (related to quantum computing)"
        ),
        "Particle types": ex['n_simples'],
        "Total dimension": f"{D_sq:.2f}",
    })

display(pd.DataFrame(rows))
```
