## The Problem

Nature distinguishes between left-handed and right-handed particles. The weak nuclear force, for example, only affects left-handed particles. This **chirality** is a fundamental feature of the Standard Model.

To compute predictions from quantum field theories on a computer, physicists place particles on a discrete grid (a "lattice"). But a collection of mathematical theorems — the **chirality wall** — says you cannot put a single left-handed particle on a lattice without automatically creating its right-handed mirror partner.

This has been an open problem for over 40 years, since Nielsen and Ninomiya proved their no-go theorem in 1981.

## Two Groups, Two Conclusions

In 2024–2026, two research groups approached this problem from different directions:

| Group | Approach | Conclusion |
|-------|----------|------------|
| **Golterman-Shamir (GS)** | Extended the original no-go theorem to interacting theories | "The wall still stands — you cannot have chiral fermions on the lattice" |
| **Thorngren-Preskill-Fidkowski (TPF)** | Proposed a new construction using extra dimensions and exotic math | "We can evade the wall using a 4+1D topological slab" |

Are they contradicting each other? Our formal verification shows: **No.** Both are correct — they are working in different mathematical spaces.

## The No-Go Rules

The Golterman-Shamir theorem says: "If your lattice system satisfies **all** of these rules, then you cannot have chiral fermions." The rules include things like:

- The grid has regular spacing (translation invariance)
- Only fermion fields are present (no auxiliary bosons)
- Interactions are short-range (local Hamiltonian)
- The particles on each grid point have a finite number of states

There are 9 rules in total. The key insight: **the no-go only works if all 9 rules hold at the same time.** Break any one, and the theorem no longer applies.

```python
conditions = gs_conditions()

rows = []
for c in conditions:
    rows.append({
        "Rule": c.name.replace("_", " ").title(),
        "In plain language": c.description[:120] + "...",
        "Does TPF obey it?": c.applies_to_tpf.value.upper(),
    })

display(pd.DataFrame(rows))
```

## The Key Result

Two groups reached different conclusions about whether chiral fermions can live on a lattice. We formally verified — using the Lean 4 theorem prover — that **both are correct**.

- **Golterman-Shamir** proved their no-go theorem within a specific mathematical framework (finite-dimensional, fermion-only, fixed-dimension lattice).
- **Thorngren-Preskill-Fidkowski** work in a *different* mathematical framework (infinite-dimensional rotors, bosonic ancillas, extra dimensions).

The GS no-go simply does not apply to the TPF construction — three of its nine premises are violated. There is no contradiction.

**Caveat:** The TPF construction depends on an unproven conjecture (the "4+1D gapped interface conjecture"). If this conjecture is eventually proven, the chirality wall falls for TPF-type constructions.

## Why It Matters

If the chirality wall can be circumvented, it opens the door to:

1. **Lattice QCD with chiral fermions** — direct simulation of the Standard Model's chiral structure on a computer, without the approximations currently required
2. **Non-perturbative chiral gauge theory** — mathematical foundation for the weak interaction beyond perturbation theory
3. **Quantum simulation** — potential realization of chiral fermions on quantum hardware using rotor-based qudit systems

The formal verification approach — using a theorem prover to adjudicate between competing theoretical claims — demonstrates a new methodology for theoretical physics: when two groups disagree, formalize both arguments and let the computer check the logic.
