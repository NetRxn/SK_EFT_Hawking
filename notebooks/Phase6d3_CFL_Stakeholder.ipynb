## Inside a neutron star

Push ordinary nuclear matter to densities far above the nuclear saturation density. Eventually, individual nucleons stop making sense — the quarks are too close together. At sufficient density, theory predicts a *new phase of matter* called **color-flavor locking** (CFL):

Three quark flavors (up, down, strange) and three color charges (R, G, B) pair up into a *diquark condensate* — like Cooper pairs in a superconductor, but in a non-abelian gauge theory. The pairing breaks the SU(3)$_{\rm color}$ × SU(3)$_L$ × SU(3)$_R$ gauge/flavor symmetry down to a diagonal SU(3). The phase has remarkable properties:
- The QCD vacuum is a color superconductor.
- All eight gluons get masses.
- A nonet of pseudo-Goldstone bosons appears, organized by the residual diagonal SU(3).

## Quark-hadron continuity

Schäfer and Wilczek (1999) made a striking observation: as you *decrease* density, the CFL phase smoothly evolves into the ordinary confined hadronic phase, *without* a thermodynamic phase transition. The two phases share enough symmetry structure to be continuously connected.

**This was puzzling.** CFL and confined QCD look completely different microscopically. How can they be the same phase?

## The Hirono-Tanizaki resolution

Hirono and Tanizaki (PRL 122, 212001, 2019; arXiv:1811.10608) showed that CFL has an *emergent* one-form $\mathbb{Z}_3$ symmetry from the diquark sector — and this emergent $\mathbb{Z}_3$ has the *same generator* as the bare-gauge $\mathbb{Z}_3$ center symmetry of confined QCD. The two phases are continuously connected because they share the same higher-form symmetry, even though the underlying physics is different.

## What this paper formalizes

Two physically independent derivations:
- **Bare-gauge derivation** (Phase 6d Wave 1): take SU(3) gauge theory, find its center, get $\omega = e^{2\pi i/3}$.
- **Diquark Cooper-pair derivation** (Hirono-Tanizaki): take the CFL diquark condensate, identify the emergent one-form symmetry, get $\omega = e^{2\pi i/3}$.

Same generator. The Lean theorem `CFL_emergent_Z3_matches_QCD_center_Z3` proves equality at the level of the closed-form expression (both reduce to the literal $e^{2\pi i/3}$); the substantive claim is the *physical-derivation independence*, not Python-level computational independence (both Python paths necessarily evaluate to the same complex number to machine precision).

## 2. Why the agreement is non-trivial

Two derivations starting from completely different physics could easily have produced different generators — say, $\zeta = e^{i\pi/3}$ (a 6th root) or $-1$ (a 2nd root). The fact that *both* arrive at the *same* primitive cube root $\omega$ is what makes quark-hadron continuity work.

The algebraic consequences of $\omega = e^{2\pi i/3}$ are:
- $\omega^3 = 1$ (cube root of unity)
- $|\omega| = 1$ (lives on the unit circle)
- $1 + \omega + \omega^2 = 0$ (this is what *distinguishes* $\mathbb{Z}_3$ from $\mathbb{Z}_2$ — for $\mathbb{Z}_2$, $1 + (-1) = 0$ is trivial)

```python
omega = EMERGENT_Z3_PHASE
print(f'  ω           = {omega}')
print(f'  ω²          = {omega**2}')
print(f'  ω³          = {omega**3}      (≈ 1: {abs(omega**3 - 1) < 1e-10})')
print(f'  |ω|         = {abs(omega):.10f}      (≈ 1: {abs(abs(omega) - 1) < 1e-10})')
print(f'  1 + ω + ω²  = {1 + omega + omega**2}      (≈ 0: {abs(1 + omega + omega**2) < 1e-10})')
print()
print('  → The 1 + ω + ω² = 0 identity is THE algebraic signature distinguishing Z_3.')
print('    Lean: emergentZ3_sum_cube_roots, derived through the correctness-push.')
```

```python
for label, m_q, phi, d_phi in [
    ('chiral limit (m_q = 0)',          0.0,    1.0+0j, 1.5),
    ('non-CFL vacuum (Φ = 0)',          0.0035, 0+0j,   0.0),
    ('CFL phase, PDG light-quark mass', 0.0035, 1.0+0j, 1.5),
    ('CFL phase, complex Φ',            0.0035, 0.6+0.8j, 1.5),
]:
    kin = cfl_kinetic_term(d_phi)
    mass = cfl_mass_term(m_q, phi)
    print(f'  {label:38s}  |∂Φ| = {d_phi:.2f}  kinetic = {kin:.4f}   mass = {mass:.6e}')
```

**Honest scope:** in this notebook (and the Lean module), `H_TopologicalOrderBeyondLG` is a *tracked Prop* — an external assumption that constrains $\mathbb{Z}_3$ charges to $\{1, 2\}$ for the topologically-ordered sector. We do *not* derive local-probe invisibility from first principles; the tracked Prop encodes Hirono-Tanizaki's claim as a hypothesis the cross-bridge consumes.

```python
for label, charge in [
    ('charge 0  (trivial vacuum, FALSIFIER)',     0),
    ('charge 1  (non-trivial topological sector)', 1),
    ('charge 2  (non-trivial topological sector)', 2),
    ('charge 3  (out of Z_3, FALSIFIER)',         3),
]:
    h = H_TopologicalOrderBeyondLG(charge=charge)
    flag = '✓ topological' if h.holds else '✗ not topological'
    print(f'  {label:48s}  {flag}')
```

## 6. Honest scope

**What this work proves.** The Phase 6d Wave 1 bare-gauge $\mathbb{Z}_3$ generator and the Phase 6d Wave 3 CFL emergent $\mathbb{Z}_3$ generator are *the same primitive cube root of unity*. The cube-roots-sum identity $1 + \omega + \omega^2 = 0$ holds, distinguishing $\mathbb{Z}_3$ from $\mathbb{Z}_2$. The CFL chiral Lagrangian's load-bearing mass-term positivity requires both $m_q > 0$ and the CFL phase. 12 Lean theorems, zero `sorry`, zero new axioms.

**What this work does not prove.**

1. *That CFL is the actual ground state of dense QCD.* Lattice QCD at finite chemical potential has a sign problem; the CFL prediction is theoretical, not lattice-confirmed.
2. *Quark-hadron continuity in full detail.* We formalize the $\mathbb{Z}_3$ matching, but not the smooth interpolation between phases.
3. *That neutron-star observations specifically test CFL.* Neutron-star equation-of-state constraints are consistent with quark cores, but cannot uniquely identify CFL vs other dense-matter phases.

**What would falsify the bridge.**

1. A reformulation of the Hirono-Tanizaki diquark-sector calculation showing the emergent generator is *not* a primitive cube root — would invalidate the matching.
2. Non-perturbative QCD work showing the CFL phase has a different higher-form symmetry structure than claimed.

**Why this matters strategically.** The $\mathbb{Z}_3$-matching is *the* Phase 6d correctness-push anchor. It ties together two waves that were developed independently (W1 from the gauge-theoretic / KSS / confinement direction, W3 from the dense-matter / Cooper-pairing direction) and shows their target structures align algebraically.
