## The Big Picture

Hawking radiation was predicted in 1974 but never observed from real black holes --- it's
billions of times too cold. "Analog" black holes in fluids produce the same effect at
accessible temperatures. Atomic BEC confirmed this in 2016 at ~0.35 nanokelvin.

**Polaritons** --- quasiparticles of light in semiconductor microcavities --- offer
temperatures 10 billion times higher (~60 millikelvin). A Paris group demonstrated
all prerequisites in 2025. This notebook presents the quantitative predictions.

## The Detection Shortcut: Stimulated Hawking

Instead of waiting for spontaneous quantum emission (requiring ~10^9 repetitions),
we send a probe laser at the horizon and measure the amplification. This is like
measuring laser gain --- classical optics rather than quantum correlation.

```python
kappa = paris['kappa']

# Gain at the optimal probe frequency
omega_opt = 0.1 * kappa
G = stimulated_hawking_gain(omega_opt, kappa)
snr = stimulated_hawking_snr(omega_opt, kappa, n_probe=100, n_shots=1)

print(f"At optimal probe frequency (omega = 0.1 * kappa):")
print(f"  Gain = {G:.2f} (probe amplified by {G*100:.0f}%)")
print(f"  SNR with just 100 probe photons: {snr:.0f}")
print(f"  That's a clear detection in a SINGLE measurement!")
```
