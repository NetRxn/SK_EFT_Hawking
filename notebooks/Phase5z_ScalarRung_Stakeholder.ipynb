## 2. The Higgs Mass from a Mexican Hat

A Mexican-hat potential has three numbers: $\mu^2$ (the negative-mass-squared parameter that drives instability), $\lambda$ (the quartic that stabilizes the potential), and the resulting VEV $v$.

The familiar SM relations are:
- $v^2 = \mu^2/\lambda$ — the field's vacuum value
- $m_H^2 = 2\lambda v^2$ — the Higgs particle mass

Setting $\mu^2 = \lambda v^2$ at $v = 246.22$ GeV (the observed EW VEV) and $\lambda = 0.129$ (from PDG-derived $m_H$ and $v$), we should recover $m_H \approx 125.25$ GeV. Let's check.

```python
v_ew = EW_PARAMS['V_EW_GEV']
lam_sm = EW_PARAMS['LAMBDA_SM_HIGGS']
mu_sq = lam_sm * v_ew**2

m_H_pred = higgs_mass_from_vev(mu_sq, lam_sm)
m_H_obs = EW_PARAMS['M_H_GEV']

print(f"With v = {v_ew:.2f} GeV and λ = {lam_sm:.4f}:")
print(f"  Tree-level m_H prediction = {m_H_pred:.3f} GeV")
print(f"  PDG 2024 observation      = {m_H_obs:.3f} GeV")
print(f"  Agreement                 = {(1 - abs(m_H_pred - m_H_obs)/m_H_obs)*100:.2f}%")
print()
print("⇒ The Mexican-hat machinery reproduces the Higgs mass at sub-percent level.")
```

## 3. The W and Z Masses from a Single VEV

The same VEV $v = 246.22$ GeV, when combined with the gauge couplings $g$ (SU(2)_L) and $g'$ (U(1)_Y), gives the W and Z masses via the Anderson-Higgs mechanism:

- $M_W = \tfrac{1}{2} g v$
- $M_Z = \tfrac{1}{2}\sqrt{g^2 + g'^2}\,v$

This is one of the most precise predictions in physics: the masses come from the same VEV that gives the Higgs its mass. Let's verify with the PDG 2024 gauge couplings.

```python
M_W_pred, M_Z_pred = anderson_higgs_w_z_masses()
M_W_obs = EW_PARAMS['M_W_GEV']
M_Z_obs = EW_PARAMS['M_Z_GEV']

print("Anderson-Higgs W/Z mass matrix consistency:")
print(f"  M_W: predicted {M_W_pred:.3f}, PDG {M_W_obs:.3f}, error {(M_W_pred-M_W_obs)/M_W_obs*100:.2f}%")
print(f"  M_Z: predicted {M_Z_pred:.3f}, PDG {M_Z_obs:.3f}, error {(M_Z_pred-M_Z_obs)/M_Z_obs*100:.2f}%")
print()
print("cos θ_W consistency (from gauge couplings vs from masses)")
ctw = cos_theta_w_consistency()
print(f"  predicted: {ctw['predicted']:.4f}")
print(f"  expected:  {ctw['expected']:.4f}")
print(f"  error:     {ctw['abs_error']:.4f}")
print()
print("⇒ The same scalar VEV reproduces W, Z, and m_H — sub-percent agreement.")
print("⇒ This is the substantive content of the structural identification.")
```

## 4. The Falsifiability Anchor: Microscopic m_H

The structural identification gives us $m_H^2 = 2\mu^2$. But the structural framing alone doesn't tell us why $\mu^2$ has the value it does. The Wetterich substrate provides a *microscopic* answer:

$$m_H \approx \sqrt{2\lambda_4} \cdot \Lambda_{\rm UV} \cdot \exp(-\pi^2 / (2 N_f G_c))$$

where the substrate parameters are:
- $\Lambda_{\rm UV}$ — the UV cutoff (substrate scale, e.g., GUT or Planck)
- $N_f$ — the count of fermion components (15 for SM without ν_R, 16 with)
- $G_c$ — a dimensionless 4-fermion coupling (order 1)
- $\lambda_4$ — the scalar-channel quartic

**The falsifiability question:** does $m_H = 125.25$ GeV come out of this formula naturally, for some natural region of (Λ_UV, G_c)? Or only along a tuned curve?

```python
# Probe a few representative substrate scales
test_cases = [
    ('EW (TeV)',        1e3),
    ('Intermediate',    1e8),
    ('GUT',             1e16),
    ('Planck',          1e19),
]
n_f = 15
g_c = 1.0
lam4 = 0.13
tol = EW_PARAMS['M_H_MATCH_TOLERANCE']

print(f"Microscopic m_H at fiducial (N_f={n_f}, G_c={g_c}, λ_4={lam4}):")
for label, lambda_uv in test_cases:
    mh = higgs_mass_from_condensate(lambda_uv, n_f, g_c, lam4)
    match = bool(scalar_rung_quantitative_match(mh, m_H_obs, tol))
    print(f"  Λ_UV = {label:>14s} ({lambda_uv:.0e} GeV) → m_H = {mh:.3e} GeV  [match: {match}]")

print()
print("⇒ At a single point (G_c=1, fixed λ_4), the prediction is many orders of magnitude off.")
print("⇒ Need to vary G_c to find the matching curve — see the figure below.")
```
