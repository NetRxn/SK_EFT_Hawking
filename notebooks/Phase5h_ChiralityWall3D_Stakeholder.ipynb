## What We Did

The Standard Model of particle physics has a deep asymmetry: its fundamental
particles (quarks, leptons) interact differently depending on whether they
spin left-handed or right-handed. This is called **chirality**, and for 50
years it has been impossible to put chiral fermions on a discrete lattice
without breaking this asymmetry.

In 2026, two breakthroughs appeared:
- **Gioia-Thorngren (GT)**: Constructed lattice fermions with exact chiral symmetry
- **Thorngren-Preskill-Fidkowski (TPF)**: Proposed a mechanism to gauge this symmetry

We formally verified the mathematical structure of both constructions,
tracking exactly what is proved vs what is conjectured.

## The 50-Year No-Go and How It Was Evaded

The Golterman-Shamir no-go theorem says: if you satisfy 9 specific conditions,
you **cannot** have chiral fermions on a lattice. The breakthrough insight is
that the TPF construction violates 3 of these 9 conditions by using bosonic
rotor degrees of freedom and an extra spatial dimension.

```python
gs = gs_condition_count()
evasion = tpf_evasion_count()

print(f"The no-go theorem has {gs['n_total']} conditions.")
print(f"TPF violates {evasion['n_violated']} of them: ({evasion['evasion_fraction']:.0%} evasion)")
print()
print("How TPF evades each violated condition:")
print("  1. Uses bosonic rotor ancillas (not pure fermion fields)")
print("  2. Uses infinite-dimensional Hilbert space per site (L^2(S^1) rotors)")
print("  3. Uses a 4+1D SPT slab (not purely 3+1D)")
```

## Why 16 Is Special

The Standard Model has exactly 16 Weyl fermion components per generation
(including the right-handed neutrino). This is not a coincidence: 16 is
the smallest number that cancels all anomalies in the Z16 classification
from topological quantum field theory.

```python
result = sm_anomaly_index(SM_FERMION_DATA, include_nu_R=True)
print(f"SM fermion components per generation (with nu_R): {result['total_components']}")
print(f"Anomaly index: {result['total_components']} mod 16 = {result['anomaly_mod16']}")
print(f"Anomaly-free: {result['anomaly_free']}")
print()
print("This means the SM automatically satisfies the '16|n' condition")
print("needed for the TPF gauging mechanism to work.")
print()
print("Proved in Lean 4: sm_anomaly_with_nu_R (Z16AnomalyComputation.lean)")
```
