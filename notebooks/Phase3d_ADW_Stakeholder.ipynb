## The Big Question

Can **gravity itself** emerge from more fundamental quantum building blocks, the way sound emerges from atoms?

The Akama-Diakonov-Wetterich (ADW) mechanism proposes that the fundamental field of gravity (the *tetrad*) is actually a composite object made of fermion pairs — similar to how Cooper pairs form in superconductors.

We tested this at "Level 2" of the gravity hierarchy: using *emergent* fermions from a lattice model (Wen's string-net condensation), not fundamental ones.

## Step 1: Emergent Fermions from a Lattice

Wen's model starts with a purely bosonic lattice system. Through a mechanism called *string-net condensation*, it produces emergent particles that behave exactly like electrons (Dirac fermions) at low energies.

On a 3D lattice, the Nielsen-Ninomiya theorem guarantees we get **4 species** of these fermions.

```python
qed = wen_coulomb_phase()
print(f"Number of emergent fermion species: {qed.N_f}")
print(f"Emergent force carrier: {qed.gauge_group} (like electromagnetism)")
```

## Step 2: The Phase Transition

When the fermion interaction is strong enough (above a critical value G_c), the system undergoes a **phase transition**: fermion pairs condense into a tetrad field, and gravity spontaneously appears.

This is directly analogous to:
- **Superconductivity**: electron pairs condense below T_c
- **Superfluid He-3**: atom pairs form a matrix-valued order parameter
- **Higgs mechanism**: the Higgs field condenses to give mass to particles

## Step 4: Does It Work with Emergent Fermions?

The mechanism works perfectly with *fundamental* fermions. But our program requires it to work with *emergent* fermions from the lattice. Here we hit four roadblocks:

```python
# viz-ref: fig_adw_structural_obstacles
from src.core.visualizations import fig_adw_structural_obstacles

fig_adw_structural_obstacles().show()
```

### What do these obstacles mean?

1. **Spin-connection gap**: The lattice model produces a U(1) force (like electromagnetism), but gravity needs an SO(3,1) force (the spin connection). These are very different mathematical objects.

2. **Grassmann-bosonic incompatibility**: The proof that this mechanism produces a well-behaved quantum theory relies on properties unique to fundamental fermions. Emergent fermions from bosonic lattices may not have these properties at the energy scale where gravity forms.

3. **Nielsen-Ninomiya doubling**: The lattice produces non-chiral fermions (equal left and right). The real world has chiral fermions (left and right are different).

4. **Cosmological constant**: The mechanism naturally predicts an enormous cosmological constant, far larger than observed.

## Bottom Line

| Question | Answer |

| Does the gap equation have a solution? | **Yes** (above critical coupling) |
| Does the solution have the right signature? | **Yes** (Lorentzian, automatically) |
| Do gravitons emerge? | **Yes** (2 massless spin-2 modes) |

This is a **qualified positive result**: the mechanism works at mean-field level, but the emergent fermion bootstrap faces structural barriers that need resolution. The vestigial metric phase (gravity without the full tetrad) may offer a more accessible intermediate target.
