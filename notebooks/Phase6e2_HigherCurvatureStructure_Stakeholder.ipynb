## The three curvature scalars

Pick a 4D spacetime. There are three independent ways to combine the curvature tensor into a scalar at second order:

| Scalar | Symbol | Interpretation |
|---|---|---|
| Ricci-scalar squared | $R^2$ | "trace squared" — pure scalar curvature |
| Ricci-squared | $R_{\mu\nu}^2$ | trace and trace-free Ricci pieces |
| Riemann-squared | $R_{\mu\nu\rho\sigma}^2$ | full curvature tensor |

In 4D the Gauss-Bonnet density $\mathcal{G} = R^2 - 4 R_{\mu\nu}^2 + R_{\mu\nu\rho\sigma}^2$ is **topological**: its integral over a closed 4-manifold gives $32\pi^2$ times the Euler characteristic. Locally it has a value, but it doesn't contribute to the equations of motion.

So at the level of physics there are only **two** independent combinations: $R^2$ and the Weyl-squared $C^2$ (the trace-free part of the curvature). This is Stelle's basis.
