## What is torsion, and why does it matter here?

Standard general relativity uses a **metric** $g_{\mu\nu}$ and
torsion-free geometry. Einstein-Cartan theory generalizes this to
include **torsion** — a kind of geometric defect that arises
naturally when the underlying field is a **tetrad** $e^a_\mu$
rather than a metric.

The ADW emergent-gravity programme works with tetrads (because
fermions need them), so torsion is structurally inevitable. The
physical question is: how big is it, and is it consistent with
observation?

Hehl-style EC theory tells us
$$\text{Torsion amplitude} = G_N \cdot \text{(spin density)}.$$

```python
from src.core.formulas import torsion_amplitude_at_cosmological_background

M_Pl = 1.221e19  # GeV
N_f = 16  # SM Dirac species
alpha_EC = 1.0  # Sakharov-Adler calibration
T_EC = torsion_amplitude_at_cosmological_background(M_Pl, N_f, alpha_EC)
print(f"Predicted torsion at natural params: |T_EC| ≃ {T_EC:.2e} GeV")
```

## How does this compare to observational bounds?

The tightest published bound is from Kostelecky, Russell, and
Tasson (PRL 100, 111102 (2008)):

$$T_{\mathrm{Kost}} = 10^{-31}\ \text{GeV}.$$

Our prediction is $\sim 2 \times 10^{-77}$ GeV — that's **46
orders of magnitude** of headroom below the bound.

For reference:
- 1 order of magnitude = factor of 10
- 46 orders of magnitude is the ratio between the size of an atom
  and the size of the visible universe, *squared*.

The natural ADW prediction is *spectacularly* below the bound.

```python
from src.einstein_cartan import torsion_observational_verdict

v = torsion_observational_verdict(M_Pl, N_f, alpha_EC)
print(f"Predicted amplitude         : {v.amplitude_gev:.2e} GeV")
print(f"Kostelecky bound             : 1e-31 GeV")
print(f"Headroom (orders of magnitude): {v.log10_margin_kostelecky:.0f}")
print(f"Verdict                      : {v.verdict_label}")
```

## Is the prediction zero, or just small?

It's *small*, not zero. This matters because a zero prediction
would be unfalsifiable; a small-but-nonzero prediction is a
genuine structural signature waiting for stronger experiments.

The key Lean theorem `torsionAmplitude_pos` formalizes this:
under positive cutoff and species count and spin density, the
predicted amplitude is **strictly positive**.

## What if observational precision improved?

Even if Lorentz-invariance experiments improved their precision
by 30 orders of magnitude, the ADW prediction would still sit
16 orders of magnitude below the new bound. This is *not* a
model that comes close to observational tension — it's a
model that lives comfortably inside the observed regime.

The closest the prediction gets to bounds is at low cutoffs
(TeV scale), but even there it sits at $\sim 10^{-45}$ GeV,
still 14 orders of magnitude below Kostelecky.
