## What Are Modular Tensor Categories?

MTCs describe the mathematics of **anyons** --- exotic particles that exist only in
two dimensions. When you braid anyons around each other, the system changes in a
way that depends only on the braid topology, not the details of the path. This
topological protection makes anyons ideal building blocks for **fault-tolerant
quantum computers**.

The Fibonacci anyon (from SU(2)_3) is **universal for quantum computation** ---
any quantum circuit can be approximated by braiding Fibonacci anyons.

## The Fibonacci Anyon: Universal for Quantum Computing

Only two particle types: vacuum (1) and the tau anyon.
The fusion rule tau x tau = 1 + tau means the dimension grows
by the golden ratio at each step --- hence the name.

```python
phi = (1 + np.sqrt(5)) / 2

print("Fibonacci anyon fusion rule:")
print(f"  tau x tau = 1 + tau")
print(f"  Quantum dimension of tau: phi = {phi:.6f} (golden ratio)")
print()
print("Hilbert space dimensions grow as Fibonacci numbers:")
print("  2 anyons: 2 states (1 or tau)")
print("  3 anyons: 3 states")
print("  4 anyons: 5 states")
print("  5 anyons: 8 states")
print("  n anyons: F(n) states  (Fibonacci sequence!)")
```
