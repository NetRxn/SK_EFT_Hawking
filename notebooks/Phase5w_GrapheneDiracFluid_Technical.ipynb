## 2. Acoustic Metric at the Horizon

```python
# Metric at horizon: g_tt should vanish
g_horizon = dirac_fluid_metric_3d(c_s, v_F, w=1.0, n=1.0)
print('G_μν at v = c_s (horizon):')
print(np.array2string(g_horizon, precision=4, suppress_small=True))
print(f'\ng_tt = {g_horizon[0,0]:.2e} (should be 0)')
print(f'g_yy = {g_horizon[2,2]:.4f} = Ω²')
print(f'Off-diagonal y: g_ty = {g_horizon[0,2]}, g_xy = {g_horizon[1,2]} (block-diagonal ✓)')
```

```python
# Metric subsonic: Lorentzian signature (det < 0)
g_sub = dirac_fluid_metric_3d(0.5*c_s, v_F, w=1.0, n=1.0)
det_sub = dirac_fluid_metric_det(0.5*c_s, v_F, w=1.0, n=1.0)
print('G_μν at v = c_s/2 (subsonic):')
print(np.array2string(g_sub, precision=6))
print(f'\ndet(G) = {det_sub:.4e} < 0  (Lorentzian ✓)')
```
